```python
import cv2
from pathlib import Path

# Define the function to resize images in a given directory
def resize_images_in_directory(directory_path, target_size):
    # Iterate through all files in the directory
    for image_path in directory_path.glob('**/*.jpg'):
        # Read the image
        image = cv2.imread(str(image_path))
        # Resize the image
        resized_image = cv2.resize(image, target_size, interpolation=cv2.INTER_AREA)
        # Overwrite the original image with the resized image
        cv2.imwrite(str(image_path), resized_image)
        # Yield the resized image path
        yield image_path

# Define the directories for train, test, and validation sets
train_dir = Path('/content/drive/MyDrive/merged_dataset (2)/train')
test_dir = Path('/content/drive/MyDrive/merged_dataset (2)/test')
val_dir = Path('/content/drive/MyDrive/merged_dataset (2)/validation')

# Define the target size for resizing
target_size = (224, 224)

# Resize images in each directory
train_generator = resize_images_in_directory(train_dir, target_size)
test_generator = resize_images_in_directory(test_dir, target_size)
val_generator = resize_images_in_directory(val_dir, target_size)

# Iterate over the generators to resize images
for image_path in train_generator:
    print(f"Resized image: {image_path} in train directory.")

for image_path in test_generator:
    print(f"Resized image: {image_path} in test directory.")

for image_path in val_generator:
    print(f"Resized image: {image_path} in validation directory.")
```

```python
import os
def image_processing(filepath):
    """
    This function processes images from a list of filepaths.

    Args:
      filepath: A list of filepaths to images.

    Returns:
      A Pandas DataFrame containing the processed image data.
    """

    labels = [str(os.path.dirname(filepath[i])).split(os.sep)[-1] \
              for i in range(len(filepath))]

    filepath = pd.Series(filepath, name='Filepath').astype(str)
    labels = pd.Series(labels, name='Label')

    # Concatenate filepaths and labels
    df = pd.concat([filepath, labels], axis=1)

    # Shuffle the DataFrame and reset index
    df = df.sample(frac=1).reset_index(drop = True)

    return df
```

```python
train_df = image_processing(train_filepaths)
test_df = image_processing(test_filepaths)
val_df = image_processing(val_filepaths)
```

```python
import matplotlib.pyplot as plt
import random

# Function to visualize random images
def visualize_random_images(image_paths, num_samples=5):
    # Randomly select num_samples images
    random_samples = random.sample(image_paths, num_samples)

    # Create subplots
    fig, axes = plt.subplots(1, num_samples, figsize=(15, 5))

    # Plot images
    for i, image_path in enumerate(random_samples):
        # Read the image
        image = cv2.imread(str(image_path))

        # Display the image
        axes[i].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        axes[i].axis('off')
        axes[i].set_title(f"Image {i+1}")

    # Show the plot
    plt.show()

# Define the directory path where your resized images are stored
# Assuming the images have been resized and resolution adjusted
dataset_folder = '/content/drive/MyDrive/merged_dataset (2)/'

# Define the directories for train, test, and validation sets
train_dir = Path(dataset_folder, 'train')
test_dir = Path(dataset_folder, 'test')
val_dir = Path(dataset_folder, 'validation')

# Get some sample image paths from each set
train_samples = list(train_dir.glob('**/*.jpg'))
test_samples = list(test_dir.glob('**/*.jpg'))
val_samples = list(val_dir.glob('**/*.jpg'))

# Visualize random sample images from each set
print("Visualizing random sample images from the train set:")
visualize_random_images(train_samples)

print("Visualizing random sample images from the test set:")
visualize_random_images(test_samples)

print("Visualizing random sample images from the validation set:")
visualize_random_images(val_samples)
```

```python
import xml.etree.ElementTree as ET
import cv2
import os
import matplotlib.pyplot as plt
import random

# Function to parse Pascal VOC XML annotations
def parse_voc_xml(xml_file):
    tree = ET.parse(xml_file)
    root = tree.getroot()

    # Initialize list to store bounding box coordinates
    bounding_boxes = []

    # Iterate over each object annotation
    for obj in root.findall('object'):
        # Extract class label and bounding box coordinates
        bbox = obj.find('bndbox')
        bounding_boxes.append((int(bbox.find('xmin').text), int(bbox.find('ymin').text),
                               int(bbox.find('xmax').text), int(bbox.find('ymax').text),
                               obj.find('name').text))

    return bounding_boxes

# Function to display annotated images
def display_images(image_paths, annotations, num_images=5):
    for image_path in random.sample(image_paths, num_images):
        image = cv2.imread(image_path)
        for xmin, ymin, xmax, ymax, class_label in annotations[image_path]:
            cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0, 255, 0), 2)
            cv2.putText(image, class_label, (xmin, ymin - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 1, cv2.LINE_AA)
        plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
        plt.axis('off')
        plt.show()

# Paths to XML annotations and image files
xml_filepath = r'C:\Users\Lenovo\Downloads\apple1\Annotations\image_25.xml'
image_filepath = r'C:\Users\Lenovo\Downloads\apple\image_25.jpg'

# Parse XML annotations for each image
annotations = {}
for xml_file in os.listdir(xml_filepath):
    annotations[os.path.join(image_filepath, xml_file[:-4] + '.jpg')] = parse_voc_xml(os.path.join(xml_filepath, xml_file))

# Display annotated images
display_images(list(annotations.keys()), annotations)
```
